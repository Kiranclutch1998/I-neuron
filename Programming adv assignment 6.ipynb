```python
# Question 1.
# You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position.
# Return the letter that was added to t.

def find_the_difference(s,t):
    s = list(s)
    t = list(t)
    diff = list(t)
    
    for i in range(len(s)):
        for j in range(len(t)):
            if s[i] == t[j]:
                diff.remove(t[j])
                break
    return diff[0]
```

```python
find_the_difference("abcd", "abcde") 
```

```python
find_the_difference("", "y") 
```

```python
find_the_difference("ae", "aea") 
```

```python
# Question 2. 
# Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. 
# Finally return the total in a list.

def count_datatypes(*args):
    type1 = [type(i) for i in args]
    count = [type1.count(int),type1.count(str),type1.count(bool),type1.count(list),type1.count(tuple),type1.count(dict)]
    return count
```

```python
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) 
```

```python
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23) 
```

```python
# Question 3.
# A Fibonacci string is a precedence of the Fibonacci series.

def fib_str(num,list1):
    result = []
    for i in range(num):
        if i == 0:
            result.append(list1[i])
        elif i == 1:
            result.append(list1[i])
        else: 
            result.append(result[-1] + result[-2])
    return result
```

```python
fib_str(5, ["e", "a"]) 
```

```python
fib_str(6, ["n", "k"]) 
```

```python
# 4.You need to find out how many ones, threes, and nines it would at least take for the number of each to 
# add up to the given integer when multiplied by one,three or nine

class ones_threes_nines:
    def __init__(self,n):
        nines = n //9
        a= n%9
        if a<3:
            ones = a
            threes = 0
        else:
            threes = a//3
            ones = a%3 
        self.answer = "nines:{}, threes:{}, ones:{}".format(nines,threes,ones)
```

```python
ones_threes_nines(10).answer
```

```python
ones_threes_nines(15).answer
```

```python
ones_threes_nines(26).answer
```

```python
# Question 5.
# Write a non-recursive function that takes an integer n and returns the Fibonacci sequence's value at index n

def fib(n):
    fibb =[0,1]
    while len(fibb)<(n+1):
        fibb.append(fibb[-1]+fibb[-2])
    return fibb[n]
```

```python
fib(6)
```

```python
fib(10)
```
